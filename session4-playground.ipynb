## Session4 - playground

- Import the necessary packages
- Load the iris dataset
- Create the observations `X` and labels `y`


- 必要なパッケージをインポートします
- アヤメのデータセットを読み込みます
- 観測値`X`とラベル`y`を作成します

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Load data
iris = np.genfromtxt('iris_data.csv', delimiter=',')

X = iris[:,1:]
y = iris[:,0]
print("X has shape:", X.shape)
print("y has shape:", y.shape)
```

## TASK 1
- Split the dataset into training and testing set
- Use 50% of the data for training


- データセットをトレーニングセットとテストセットに分割してください
- データの50％をトレーニングに使用してください

```python
## YOUR CODE HERE
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.5)

print("Training set size:", X_train.shape[0])
print("Testing set size:", X_test.shape[0])
```

```python
print("training set:")
print("Iris setosa:", np.sum(y_train == 0))
print("Iris versicolor:", np.sum(y_train == 1))
print("Iris virginica:", np.sum(y_train == 2))
```

```python
print("testing set:")
print("Iris setosa:", np.sum(y_test == 0))
print("Iris versicolor:", np.sum(y_test == 1))
print("Iris virginica:", np.sum(y_test == 2))
```

## TASK 2

- Standardize the observations in the training and test set


- トレーニングセットとテストセットの観測を標準化してください

```python
## YOUR CODE HERE
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#training
scaler.fit(X_train);

X_train_raw = X_train
X_train = scaler.transform(X_train_raw)
```

```python
print(np.mean(X_train))
print(np.std(X_train))
```

```python
#test
X_test_raw = X_test
X_test = scaler.transform(X_test_raw)
```

```python
print(np.mean(X_test))
print(np.std(X_test))
```
