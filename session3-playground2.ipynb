## TASK 1
On the same graph plot a histogram of sepal length for:
- the iris setosa in red,
- the iris versicolor in blue,
- the iris virginica in green.

一つのグラフにがくの長さのヒストグラムをプロットしてください。
- ヒオウギアヤメを赤い線で
- ブルーフラッグを青い線で
- バージニカを緑色の線で

---
Summary of the data structure / データ構造の概要

     1.花のタイプ
     2.がく片の長さ
     3.がく片の幅
     4.花びらの長さ
     5.花びらの幅

-。
     -アイリスセトサの場合は50（0）
     -アイリスバーシクラー用50（1）
     -50はアイリスヴァージニカ（2）

- Each row corresponds to one data point and contains five values / 各行は1つのサンプルに対応し、5つの値が含まれます:
    1. flower type / 花の種類
    2. sepal length / がくの長さ
    3. sepal width / がくの幅
    4. petal length / 花びらの長さ
    5. petal width / 花びらの幅
- Each of the species has 50 data points / それぞれの花の種類には50のサンプルがあります
    - 0 = iris setosa / ヒオウギアヤメ
    - 1 = iris versicolor / ブルーフラッグ
    - 2 = iris virginica / バージニカ

```python
## YOUR CODE HERE
plt.hist(iris[:50,1], bins = np.linspace(2, 8, 20), histtype='step', color=[1, 0, 0], label="iris setosa")
plt.hist(iris[50:100,1], bins = np.linspace(2, 8, 20), histtype='step', color=[0, 0, 1], label="iris versicolor")
plt.hist(iris[100:,1], bins = np.linspace(2, 8, 20), histtype='step', color=[0, 1, 0], label="iris virginica")
plt.xlabel("length [cm]")
plt.ylabel("count")
plt.title("Histogram of sepal ")
# the names shown in legend are set using the parameter `label` in the hist function (similarly for functions like plot)
# 凡例に表示される名前は、hist関数の`label`パラメーターで設定されています（plotなどの関数は同様です）
plt.legend()
```

It is often useful to plot 2 or more graphs in the same plot. That can be done using the function `subplot`.

Look up its usage and make a plot like this:

- On the right: a histogram of the iris versicolor sepal width.

同じプロットに2つ以上のグラフをプロットすると便利な場合があります。そういう時は`subplot`関数が使えます。
<br>
その関数の使用法を調べ、次のプロットを作成します。
- 左側：ブルーフラッグのがくの幅のプロット。
- 右側：ブルーフラッグのがくの幅のヒストグラム。

```python
## YOUR CODE HERE
plt.subplot(1,2,1)
plt.plot(iris[iris[:,0] == 1, 2], color=[0,0,1])
plt.xlabel("samples")
plt.ylabel("sepal width[cm]")
plt.title("plot")

plt.subplot(1,2,2)
plt.hist(iris[iris[:,0] == 1, 2], bins = np.linspace(2, 4, 12), color=[0, 0, 1], label="iris versicolor")
plt.xlabel("sepal width[cm]")
plt.ylabel("count")
plt.title("histogram")

plt.subplots_adjust(wspace=0.4)
plt.suptitle('sepal width of the iris versicolor ')
plt.show()
```
